```python
import numpy as np #numerical ops
import pandas as pd #read+analyze data
import matplotlib.pyplot as plt #plot graphs
from sklearn.model_selection import train_test_split #training+testing model
from sklearn.linear_model import LinearRegression #implementing linear regression

#load data from CSV file
data = pd.read_csv('home_dataset.csv') #pd.read_csv() to read dataset

#extract features and target variables
house_sizes = data['HouseSize'].values #corresponds to column names
house_prices = data['HousePrice'].values

#plt to plot data, plt.scatter() for scatter plotting price vs size
plt.scatter(house_sizes, house_prices, marker = 'o', color = 'blue')
plt.title('House Prices vs House Size')
plt.xlabel('House Size (sq.ft.)')
plt.ylabel('House Price ($ (USD))')
plt.show()
#up to here, this code yields a basic scatter graph

#train-test splits generated data into training and testing sets
#simulate show model works with new data

#split code into 80/20, for xtrain+test/ytrain+test
x_train, x_test, y_train, y_test = train_test_split(house_sizes, house_prices, test_size = 0.2, random_state = 42)

#reshaoe the data for NumPy
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)

#create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

#predict prices for the test set
predictions = model.predict(x_test)

#visualize predictions from the model
plt.scatter(x_test, y_test, marker = 'o', color = 'blue', label = 'Actual Prices')
plt.plot(x_test, predictions, color = 'pink', linewidth = 2, label = 'Predicted Prices')
plt.title('Dumbo Property Price Prediction with Linear Regression')
plt.xlabel('House Size (sq.ft.)')
plt.ylabel('House Price (millions $ (USD))')
plt.legend()
plt.show()

```
